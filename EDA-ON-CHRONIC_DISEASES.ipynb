```python
df['rc']=pd.to_numeric(df['rc'],errors='coerce')
```

```python
df.info()
```

```python
cats_col=[col for col in df.columns if df[col].dtype =='object']
nums_col=[col for col in df.columns if df[col].dtype !='object']
```

```python
for col in cats_col:
    print(f"{col} has {df[col].unique()} values \n")
```

```python
#replace unwanted values
df['dm'].replace(to_replace={' yes':'yes', '\tyes': 'yes','\tno':'no'},inplace=True)
```

```python
df['dm'].unique()
```

```python
df['cad']=df['cad'].replace(to_replace={'\tno': 'no'})
```

```python
df['cad'].unique()
```

```python
df['classification'].replace(to_replace={'ckd\t': 'ckd',},inplace=True)
```

```python
df['classification'].unique()
```

```python
df['classification']=df['classification'].map({'ckd':0,'notckd':1})
```

```python
cols=['age','bp','bgr','sc','classification']

g = sns.PairGrid(df[cols],hue='classification',palette='coolwarm')
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot,cmap='Blues_d')
g.map_diag(sns.histplot)
g.add_legend()
plt.title("PairGrid for selected columns")
plt.show()
```
